Simulation sur $n$ marché
--

```python
class OptimiseurDepense(object):
    
    def __init__(self,data,denre,quantite):
        
        self.data = data
        self.denre = denre
        self.quantite = quantite
        self.achat = [(c,j) for c,j in zip(self.denre,self.quantite)]
        
        
    def make_course(self,):


        self.stock = {}        # variable qui permet de stocker la somme totale des achats par marché
        self.comp = []         # variable qui permet de stocker la somme totale des achats
        self.marches = {}      # variable qui permet de stocker les differentes facture etablit par marché

        for mkt in list(self.data.keys()):
        
            cle,val=zip(*self.data[mkt].items())
            total = []    # variable pour stocker la somme total des depanse d'achat
            p = []        # variable qui stock le prix unitaire des denré
            prod = []     # variable qui stock les differente denré
            qt = []       # variable qui stock les differente quantité
            prix = 0   
            fact = {}     # variable qui permet de stocker les differents elements de facture
            for c in self.achat:
                for i in range(len(val)):
                    for j in val[i]:
                        if c[0] not in j:
                            pass
                        else:
                            prix = j[1]
                            p.append(prix)
                            total.append(prix*c[1])
                            prod.append(c[0].upper())
                            qt.append(c[1])
                        
        # dictionnaire pour la construction des facture par marché
                            fact['Produits'] = prod
                            fact['Quantité'] = qt
                            fact['Prix unitaire'] = p
                            fact['Achat'] = total
    
        # dictionnaire permettant d'associer les differente facture au different marché
                            self.marches[mkt] = fact
    
        # construction dictionnaire de depense totale par marché
            self.stock[sum(total)] = mkt
        
        # liste des depense totale par marché pour l'identification des depense optimale
            self.comp.append(sum(total))
  
    
    def get_facture(self,):
        
        def optimisation_depense():
            minimum = 0
            for i in self.comp:
                if i == np.min(self.comp):
                    minimum = i
            return minimum 
    
        df = self.marches[self.stock[optimisation_depense()]]
        import datetime
        today = datetime.date.today()
        facture = pd.DataFrame(df)

        print('='*68)
        print("{}  {:>56}".format("Facture","date: "+str(today)))
        print('-'*68)
        print(facture)
        print("="*68)
        print("Montant total :{:>50}".format(np.sum(pd.to_numeric(facture['Achat']))))
        
        
```

```python
import streamlit as st
import mysql.connector as mc
from PIL import Image
import warnings
warnings.filterwarnings('ignore')
import datetime

image = Image.open('Desktop/IvoireEasy.png')


#Contruction de la connexion à la base de données


def connexion():
    
    try:
        db = mc.connect(host = "127.0.0.1",
                        user = "root",
                        password = "",
                        database = "IvoirEasy")
        print("succès")
        return db
    
    except:
        print("connexion non établit")
        
        
def ajout_donnees(NomVille,NomMarche,NomProduit,PrixProduit,Date):
    db = connexion()
    querry = db.cursor()
    valeur = (NomVille,NomMarche,NomProduit,PrixProduit,Date)
    querry.execute("""
                         INSERT INTO ie_db (NomVille,NomMarche,NomProduit,PrixProduit,Date)
                         VALUE (%s,%s,%s,%s,%s)
                         """,valeur)
                         
    db.commit()
    querry.close()
    
    

#Construction du formulaire de collect de données


def apply():
    st.title("IvorEasy DataBase")
    st.image(image, use_column_width=True)
    
    # creation du menu
    menu = ["Acceuil","Qui sommes nous"]
    choix = st.sidebar.selectbox("Menu", menu)
    
    # creation du formulaire
    if choix == 'Acceuil':
        st.subheader("Formulaire de collect de données")
        
        with st.form(key = 'form'):
            
            ville = st.text_input("Nom de la ville")
            marche = st.text_input("Nom du marché")
            prod = st.text_input("Nom du produit")
            prix = st.number_input("Prix du produit")
            
            enregistrement = st.form_submit_button(label="Enregistrer")
            
        if enregistrement:
            today = datetime.date.today()
            try:
                ajout_donnees(ville, marche, prod, prix,today)
                st.success("Les données ont bien été enregistrer")
            
            except:
                print("Probleme avec la logique du code")
    
            
           
          
    
       
if __name__ == '__main__':
    apply()
```
